El sistema no tiene energ√≠a potencial porque se desplaza sobre la superficie del agua.


Las varibles generalizadas son $x$, $y$ y $\theta$, de forma que el Lagrangiano del sistema es:

$$
\dfrac{d}{dt} \left( \dfrac{\partial L}{\partial \dot{q}_i} \right) - \dfrac{\partial L}{\partial q_i} = f_i

Para la variable $x$:

$$

$$

Para la variable $y$:

$$

Y finalmente para la variable $\theta$, a sabiendas de que las fuerzas realizan sendos momentos en el catamar√°n $\tau_1 = -F_1 \cdot b$ y $\tau_2 = F_2 \cdot b$, donde $b$ es la distancia ortogonal de los motores al eje de simetr√≠a, el Lagrangiano es:
